**Question 01**: How does the correlation between features change in the top 10 countries with the LARGEST population?

When we have an answer for this question, we might see how features interaction with each other when the country population is large.

```python
# PRE-PROCESSING IF NEEDED
# YOUR CODE HERE (OPTION)
def population_countries(df, is_ascending):
    # create a new dataframe
    top_10_popu_df = pd.DataFrame()
    # Find average population for each country
    top_10_popu_df["Mean population"] = df.groupby("Country")["Total Population"].mean()
    # Set the index as a column - countries
    top_10_popu_df["Country"] = top_10_popu_df.index
    # Drop the index
    top_10_popu_df.reset_index(drop=True, inplace=True)
    # Sort the rows based of Average Birth Rate
    top_10_popu_df.sort_values("Mean population", inplace=True, ascending=is_ascending)
    # Display
    return list(top_10_popu_df.head(10)['Country'])
```

```python
# YOUR CODE HERE
countries = population_countries(cleaned_df, False)
print(countries)
top_population_df = cleaned_df[cleaned_df['Country'].isin(countries)]
# Exclude the categorical features from the matrix
top_population_df.drop(['Year', 'Country'], inplace=True, axis='columns')

# plot a correlation matrix
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(top_population_df.corr(), cmap='RdBu', center=0,ax=ax)
plt.show()
```

**YOUR ANSWER**
- Observing the `correlation matrix between features in the top 10 countries with the largest population`, we can see that the correlation between most features has significantly **increased**. If initially the features were positively correlated, now this positive correlation is **even more pronounced** (transitioning from light blue to dark blue). Similarly, if the features were initially negatively correlated, now this negative correlation has also become **more pronounced**, as evidenced by the transition from light orange to brown red. This indicates that features have a stronger correlation with each other in countries with larger populations.
- However, there are still a **few features** that have slightly **decreased correlation**  or transitioned from positive to negative correlation, but the changes are **not significant**.

- We determine that the development of a country is reflected through features such as `National Income Capita, Employment in industry/agriculture, electric/energy consumtion`.
- We build a correlation matrix to see how these features affect `Average life expentancy`.
- From there, we can deduce how the development of a country affects the life expectancy of its people.

### Pre-processing

```python
# YOUR CODE HERE (OPTION)
```

### Analyzing to get the answer

#### Calculate the average life expectancy of a population

```python
features_1 = ['Employment in Industry(%)','Employment in Agriculture(%)', 'National Income per Capita',
              'Electric Power Consumption(kWH per capita)','Renewable Energy Consumption (%)',
              'Fossil Fuel Consumption (%)','Male life expectancy','Female life expectancy ']
df_1 = cleaned_df[features_1]
# Combine male/female life expectancy into one column - life expectancy
df_1['Average life expectancy'] = (df_1['Female life expectancy '] + df_1['Male life expectancy'])/2
df_1.drop(['Female life expectancy ', 'Male life expectancy'], inplace=True, axis='columns')
# Move `Average life expectancy` to the first column
column = df_1.pop('Average life expectancy')
df_1.insert(0, 'Average life expectancy', column)
# display
df_1.head()
```

#### Build a correlation matrix

```python
# YOUR CODE HERE
# plot a correlation matrix
fig, ax = plt.subplots(figsize=(10,10))
sns.heatmap(df_1.corr(), annot=True, cmap='coolwarm', center=0,ax=ax)
plt.show()
```

- From the correlation matrix, we observe that:
    - `Electric Power Consumption and Renewable Energy Consumption` have a **positive correlation** with `Average life expectancy`, indicating that they have a positive influence on the life expectancy of the population.
    - However, other factors such as `Employment in Industry/Agriculture, National Income per Capita, and Fossil Fuel Consumption` have a **negative correlation** with `Average life expectancy`, indicating how the **adverse effects** of industrial development impact the health and life expectancy of the population.

## Making your own question 02? (1.5 points)


**Question:** How is the gender distribution in the most populous countries in the recent years in Europe ?

**Answering this question will** helps you understand the gender balance within the population and can reveal different patterns and trends among populous countries.

**How we answer this question**: 
- We calculate the male ratio in the most populous countries in Europe and then plot a bar chart to visualize the gender distribution in these countries.

### Pre-processing

```python
# YOUR CODE HERE (OPTION) 
```

### Analyzing to get the answer

#### Calculate male ratio in the most populous countries in Europe

```python
# create new df
df_2 = pd.DataFrame()
# Find 10 populous countries in Europe
populous_countries = population_countries(cleaned_df, False)
temp_df = cleaned_df[cleaned_df['Country'].isin(populous_countries)]
# Find average population for each country
df_2['Avg Total Population'] = temp_df.groupby("Country")["Total Population"].mean()
df_2['Avg Male Population'] = temp_df.groupby("Country")["Male Population"].mean()
# Set the index as a column - countries
df_2["Country"] = df_2.index
# Drop the index
df_2.reset_index(drop=True, inplace=True)
# Calculate sex ratio
df_2['Sex Ratio'] = df_2["Avg Male Population"]/df_2['Avg Total Population']
# display
df_2
```

#### Plot the bar chart

```python
# YOUR CODE HERE
plt.figure(figsize=(8, 5))
sns.set(style="whitegrid")
ax = sns.barplot(data=df_2, x="Country", y= "Sex Ratio")
# Show values on the top of each bar
for index, row in df_2.iterrows():
    value = round(row['Sex Ratio'], 3)
    ax.annotate(str(value), xy=(index, row['Sex Ratio']), ha='center', va='bottom')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.show()
```

From the above bar chart, we can see that:
- The gender ratios in 9 out of the top 10 most populous countries in Europe tend to be **balanced**. This balance is not surprising as these countries are mostly highly developed.
- It demonstrates the effectiveness of development policies in these countries, particularly `gender-related policies`.
- However, `Ukraine` has the most imbalanced gender ratio, with a higher proportion of females than males by around **10%**.
